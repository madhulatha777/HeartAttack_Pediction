```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv('heart_attack_prediction_dataset.csv')
df.head()
```

```python
df[['up', 'down']] = df['Blood Pressure'].str.split('/', expand=True)
```

```python
df.head()
```

```python
df.drop(columns=['Blood Pressure'], inplace=True)
```

```python
df.head()
```

```python
df = df.drop(columns=['Patient ID', 'Country', 'Continent', 'Hemisphere'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['up'] = df['up'].astype(float)
df['down'] = df['down'].astype(float)
```

```python
X = df.drop(columns=['Heart Attack Risk'])
y = df['Heart Attack Risk']
```

```python
y.value_counts()
```

```python
y.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
ohe_columns = [columns for columns in X.columns if X[columns].dtype=='O']
numerical_columns = [columns for columns in X.columns if X[columns].dtype!='O' and columns!='Blood Pressure']
```

```python
print(ohe_columns)
print(numerical_columns)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

numeric_transformer = StandardScaler()
one_hot_encoder = OneHotEncoder(drop='first')



preprocessor = ColumnTransformer(
    [
        ("StandardScaler", numeric_transformer, numerical_columns),
        ("OneHotEncoder", one_hot_encoder, ohe_columns)
    ],
    remainder='passthrough'
)
```

```python
X = df.drop(columns=['Heart Attack Risk'])  # Keep it a DataFrame
y = df['Heart Attack Risk']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Now works fine with column-based transformers
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
import pickle
import os

# Save preprocessor
with open('preprocessor.pkl', 'wb') as f:
    pickle.dump(preprocessor, f)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
```

```python
scale_pos_weight = 5624 / 3139
weights = {
    0: 1.0,
    1: 1.79
}
models = {
    'XGBoost' : XGBClassifier(scale_pos_weight=scale_pos_weight, n_estimators=100),
    "Adaboost": AdaBoostClassifier(),
    "Random Forest": RandomForestClassifier(class_weight=weights, n_estimators=100),
    "Decision Tree": DecisionTreeClassifier(class_weight=weights),
    "Logistic": LogisticRegression(class_weight=weights, max_iter=1000),
}
trained_models = {}
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    ## Training set performance
    model_train_accuracy = accuracy_score(y_train,y_train_pred)
    model_train_f1 = f1_score(y_train,y_train_pred,average='weighted')
    model_train_precision = precision_score(y_train,y_train_pred)
    model_train_recall = recall_score(y_train,y_train_pred)
    model_train_rocauc_score = roc_auc_score(y_train,y_train_pred)

    ## Testing set performance
    model_test_accuracy = accuracy_score(y_test,y_test_pred)
    model_test_f1 = f1_score(y_test,y_test_pred,average='weighted')
    model_test_precision = precision_score(y_test,y_test_pred)
    model_test_recall = recall_score(y_test,y_test_pred)
    model_test_rocauc_score = roc_auc_score(y_test,y_test_pred)


    print(list(models.keys())[i])

    print("Model Performance for Training set")
    print("- Accuracy: {:.4f}".format(model_train_accuracy))
    print("- F1 score: {:.4f}".format(model_train_f1))
    print("- Precision: {:.4f}".format(model_train_precision))
    print("- Recall: {:.4f}".format(model_train_recall))
    print("- ROC AUC Score: {:.4f}".format(model_train_rocauc_score))


    print('------------------------------------------------------------------')

    print("Model Performance for Test set")
    print("- Accuracy: {:.4f}".format(model_test_accuracy))
    print("- F1 score: {:.4f}".format(model_test_f1))
    print("- Precision: {:.4f}".format(model_test_precision))
    print("- Recall: {:.4f}".format(model_test_recall))
    print("- ROC AUC Score: {:.4f}".format(model_test_rocauc_score))

    trained_models[list(models.keys())[i]] = model

    print()
```

```python
# Save XGBClassifier
with open( 'xgboost_model.pkl', 'wb') as f:
    pickle.dump(trained_models['XGBoost'], f)
```

```python
ann = tf.keras.models.Sequential()
```

```python

ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
from sklearn.utils import class_weight
import numpy as np
```

```python
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights_dict = dict(enumerate(class_weights))
```

```python
ann.fit(X_train, y_train, epochs=100, batch_size=32, class_weight=class_weights_dict)
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.4)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print(confusion_matrix(y_test, y_pred))
print(f'Accuracy score: {accuracy_score(y_test, y_pred)}')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=32, class_weight=class_weights_dict)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print(confusion_matrix(y_test, y_pred))
print(f'Accuracy score: {accuracy_score(y_test, y_pred)}')
```
